```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import cartopy.crs as ccrs
```

**Check spatial distrubution of grid**

```python
path = "data/EECRA/ocean_seasonal/"
df = pd.read_csv(path+'df_JJA52.csv')
lat = df['Lat']
lon = df['Lon']
```

```python
#plot count obsv gridded

# Convert 0–360 to -180–180
lon = ((lon + 180) % 360) - 180

# Bin edges for 5°x5° grid (adjust range if needed)
lat_bins = np.arange(-90, 91, 5)
lon_bins = np.arange(-180, 181, 5)

counts, lat_edges, lon_edges = np.histogram2d(lat, lon, bins=[lat_bins, lon_bins])

lon_grid, lat_grid = np.meshgrid(lon_edges, lat_edges)

fig, ax = plt.subplots(figsize=(12, 6), subplot_kw={'projection': ccrs.PlateCarree(central_longitude=180)})

mesh = ax.pcolormesh(
    lon_grid,
    lat_grid,
    counts,
    cmap='hot_r',
    shading='auto',
    transform=ccrs.PlateCarree()
)

# Add coastlines and gridlines
ax.coastlines(linewidth=0.5)
gl = ax.gridlines(draw_labels=True, linewidth=0.3, color='gray', alpha=0.7, linestyle='--')
gl.right_labels = False
gl.top_labels = False

# Add colorbar
cbar = plt.colorbar(mesh, ax=ax, orientation='vertical', pad=0.03, aspect=30)
cbar.set_label("Observation Count")

# Add title
ax.set_title("EECRA 1952 JJA 5°x5° Gridded Count of Observations", fontsize=14)

plt.tight_layout()
plt.show()
```

```python
df_jja65 = pd.read_csv(path+"df_JJA65.csv")
df_jja65_fog = df_jja65.loc[df_jja65["CL"] == 11]
lat = df_jja65_fog['Lat']
lon = df_jja65_fog['Lon']
```

```python
# Convert 0–360 to -180–180
lon = ((lon + 180) % 360) - 180

# Bin edges for 5°x5° grid (adjust range if needed)
lat_bins = np.arange(-90, 91, 5)
lon_bins = np.arange(-180, 181, 5)

counts_fog, lat_edges, lon_edges = np.histogram2d(lat, lon, bins=[lat_bins, lon_bins])

counts_probOccurance = counts_fog.astype("float") / counts

lon_grid, lat_grid = np.meshgrid(lon_edges, lat_edges)

fig, ax = plt.subplots(figsize=(12, 6), subplot_kw={'projection': ccrs.PlateCarree(central_longitude=180)})

mesh = ax.pcolormesh(
    lon_grid,
    lat_grid,
    counts_probOccurance,
    cmap='hot_r',
    shading='auto',
    transform=ccrs.PlateCarree()
)

# Add coastlines and gridlines
ax.coastlines(linewidth=0.5)
gl = ax.gridlines(draw_labels=True, linewidth=0.3, color='gray', alpha=0.7, linestyle='--')
gl.right_labels = False
gl.top_labels = False

# Add colorbar
cbar = plt.colorbar(mesh, ax=ax, orientation='vertical', pad=0.03, aspect=30)
cbar.set_label("Observation Count")

# Add title
ax.set_title("EECRA 1952 JJA 5°x5° Gridded Count of Observations", fontsize=14)

plt.tight_layout()
plt.show()
```
