```python
path = "data/EECRA/ocean/"
df2 = pd.concat(map(pd.read_csv, [path + 'df_JUN52O.csv', path + 'df_JUL52O.csv', path + 'df_AUG52O.csv']))
```

```python
df2_fog = df2.where(df2["CL"] == 11)
df2_CL = df2.where(df2["CL"] != pd.NA)
```

```python
pointPlot = plt.scatter(df2.Lon, df2.Lat, s=1, c= ("g", 0.1), label = "")
pointPlot = plt.scatter(df2_CL.Lon, df2_CL.Lat, s=1, c= ("y", 0.1), label = "")
pointPlot = plt.scatter(df2_fog.Lon, df2_fog.Lat, s=1, c= ("b", 0.4), label = "fog observed")
plt.title("")
```
