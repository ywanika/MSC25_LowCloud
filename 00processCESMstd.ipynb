```python
import xarray as xr
import os
from datetime import timedelta
```

This file breaks down cleaning and isolating the CLOUD variable for CESM std. It also creates seperaate month and yr cols. The new data is saved to cloud_std.nc.

```python
path = '//Users/anka/Desktop/research/msc25/data/cesm_StdModel'
file = 'b.e11.B20TRC5CNBDRD.f09_g16.002.cam.h0.CLOUD.192001-200512.nc'
```

```python
ds = xr.open_dataset(os.path.join(path, file))
ds['time'] = ds.time.get_index('time') - timedelta(days=16) # correct time in the file, it should starts from Jan
ds = ds.convert_calendar('standard')
ds
```

```python
cloud_std = ds.CLOUD.isel(lev=-1) # select the last level as fog
cloud_std
```

```python
cloud_std = cloud_std.assign_coords(month=("time", cloud_std['time'].dt.month.data))
cloud_std = cloud_std.assign_coords(year=("time", cloud_std['time'].dt.year.data))
cloud_std
```

```python
cloud_std.to_netcdf(path="data/cesm_StdModel/cloud_std.nc")
```
