**DJF 1972, marine fog obsv**

```python
path = "data/EECRA/ocean_seasonal/"
df3 = pd.read_csv(path+"df_DJF1972.csv")

df3_fog = df3.loc[df3["CL"] == 11]
df3_CL = df3.loc[df3["CL"].notna()]
```

```python
figure, ax = plt.subplots(figsize=(8, 5), subplot_kw={'projection': ccrs.PlateCarree(central_longitude=0)})

pointPlot = plt.scatter(df3_CL.Lon, df3_CL.Lat, s=1, c= ("y", 0.1), label = "point measuring low cloud")
pointPlot = plt.scatter(df3_fog.Lon, df3_fog.Lat, s=2, c= ("b", 0.1), label = "fog observed")
plt.legend(markerscale=6)
plt.title("EECRA DJF 1972 Fog observed")

ax.coastlines(linewidth=0.5)

plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.show()
```

**1952/3 DJF**

```python
djf53 = pd.read_csv(path+"df_DJF1953.csv")

djf53_fog = djf53.loc[djf53["CL"] == 11]
djf53_CL = djf53.loc[djf53["CL"].notna()]
```

```python
figure = plt.figure(figsize=(8, 5))

pointPlot = plt.scatter(djf53.Lon, djf53.Lat, s=1, c= ("g", 1), label = "data point")
pointPlot = plt.scatter(djf53_CL.Lon, djf53_CL.Lat, s=1, c= ("y", 1), label = "point measuring low cloud")
pointPlot = plt.scatter(djf53_fog.Lon, djf53_fog.Lat, s=2, c= ("b", 0.1), label = "fog observed")
plt.legend(markerscale=6)
plt.title("EECRA 1953 DJF Fog observed")

plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.show()
```

```python
def shift_lon_center(lon_array):
    return ((lon_array - 240 + 180) % 360) - 180 + 240

djf53['Lon_shifted'] = shift_lon_center(djf53['Lon'])
djf53_CL['Lon_shifted'] = shift_lon_center(djf53_CL['Lon'])
djf53_fog['Lon_shifted'] = shift_lon_center(djf53_fog['Lon'])

figure = plt.figure(figsize=(8, 5))

plt.scatter(djf53['Lon_shifted'], djf53['Lat'], s=1, c='g', label="data point")
plt.scatter(djf53_CL['Lon_shifted'], djf53_CL['Lat'], s=1, c='y', label="point measuring low cloud")
plt.scatter(djf53_fog['Lon_shifted'], djf53_fog['Lat'], s=2, c='b', alpha=0.1, label="fog observed")

plt.legend(markerscale=6)
plt.title("EECRA 1953 DJF Fog observed")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.xlim(60, 420)  # Optional: Set limits to show full 360 range centered on 240
plt.xticks(np.arange(60, 421, 60))  # Optional: Customize ticks

plt.show()
```
